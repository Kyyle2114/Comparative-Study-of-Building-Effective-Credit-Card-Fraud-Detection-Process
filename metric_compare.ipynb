```python
xgb_result = pd.read_csv('./compare_result/xgb_iter_test.csv', index_col=0)
lof_result = pd.read_csv('./compare_result/lof_iter_test.csv', index_col=0)
ae_result = pd.read_csv('./compare_result/ae_iter_test.csv', index_col=0)
ocsvm_result = pd.read_csv('./compare_result/ocsvm_iter_test.csv', index_col=0)
iso_result = pd.read_csv('./compare_result/iso_iter_test.csv', index_col=0)
```

```python
result_list = [xgb_result, lof_result, ae_result, ocsvm_result, iso_result]

df = pd.DataFrame()

for result in result_list:
    df = pd.concat([df, result['F1_all']], axis=1)
    
df.columns = ['XGBoost', 'LOF', 'AutoEncoder', 'OCSVM', 'Isolation Forest']

df
```
